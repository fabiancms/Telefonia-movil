# 📱 Generación de Líneas Móviles

En esta sección se generan las líneas móviles que harán parte del entorno simulado de telecomunicaciones.

Cada línea representa un usuario móvil dentro de la red y contará con información básica como:

- Identificador único de línea
- Número telefónico
- Nombre de usuario
- Área o departamento empresarial
- Perfil de consumo

Adicionalmente, se simularán diferentes comportamientos de uso mediante perfiles de tráfico:

- **High** → usuarios con alto consumo de llamadas
- **Medium** → usuarios con consumo promedio
- **Low** → usuarios con bajo consumo

Esta segmentación permitirá generar posteriormente patrones de tráfico más realistas y análisis de comportamiento dentro de la red móvil.

---

````python
# ==================================================
# GENERACIÓN DE LÍNEAS MÓVILES
# ==================================================

# Departamentos simulados
departments = ['Ventas', 'Soporte', 'IT', 'Finanzas']

# Lista para almacenar líneas móviles
lines_data = []

# Diccionario para almacenar perfiles de consumo
line_profiles = {}

# Generación de líneas
for i in range(1, 301):

    # Asignación probabilística de perfiles
    profile = random.choices(
        ['high', 'medium', 'low'],
        weights=[0.2, 0.5, 0.3]
    )[0]

    # Guardar perfil
    line_profiles[i] = profile

    # Crear registro de línea
    lines_data.append({
        "line_id": i,
        "phone_number": f'300{random.randint(1000000,9999999)}',
        "user_name": f'User_{i}',
        "department": random.choice(departments),
        "profile": profile
    })

# Crear DataFrame
lines_df = pd.DataFrame(lines_data)

# Mostrar primeras filas
lines_df.head()
````

## ✅ Análisis

Se generaron exitosamente las líneas móviles que harán parte del entorno simulado de telecomunicaciones.

Cada línea posee un perfil de comportamiento que influirá directamente en la cantidad de tráfico generado posteriormente.

La distribución de perfiles fue diseñada de forma probabilística para representar un comportamiento más cercano a un entorno real:

- Una pequeña proporción de usuarios genera alto tráfico
- La mayoría presenta un consumo promedio
- Algunos usuarios tienen bajo nivel de actividad

Este enfoque permitirá construir análisis más realistas sobre utilización de red, congestión y comportamiento de usuarios.

---

````python
# ==================================================
# DISTRIBUCIÓN DE PERFILES
# ==================================================

# Conteo de perfiles
profile_distribution = lines_df['profile'].value_counts()

print(profile_distribution)

# Gráfico de distribución
plt.figure(figsize=(8,5))

sns.countplot(
    data=lines_df,
    x='profile',
    order=['high', 'medium', 'low']
)

plt.title('Distribución de Perfiles de Consumo')
plt.xlabel('Perfil')
plt.ylabel('Cantidad de Usuarios')

plt.show()
````

## 📊 Análisis de Distribución de Perfiles

La distribución de perfiles evidencia un comportamiento controlado y alineado con escenarios reales de telecomunicaciones.

Se observa que:

- Los usuarios de perfil **Medium** representan la mayor parte de la población.
- Los perfiles **High** son menos frecuentes, pero posteriormente generarán un volumen considerable de tráfico.
- Los perfiles **Low** representan usuarios con baja actividad dentro de la red.

Este tipo de segmentación es común en operadores móviles, donde un pequeño porcentaje de usuarios suele concentrar gran parte del consumo de red.

# 🗼 Generación de Estaciones Base

En esta sección se simularán las estaciones base que conforman la infraestructura de la red móvil.

Las estaciones base permiten la comunicación entre los dispositivos móviles y la red del operador, siendo fundamentales para garantizar cobertura, capacidad y calidad del servicio.

Para este proyecto se generarán estaciones base distribuidas en diferentes ciudades de Colombia, incluyendo zonas y sectores representativos dentro de cada ciudad.

Cada estación base contará con:

- Identificador único
- Nombre de estación
- Ciudad asociada

Esta información permitirá posteriormente realizar análisis geográficos, distribución de tráfico y evaluación de desempeño de red por región.

---

````python
# ==================================================
# GENERACIÓN DE ESTACIONES BASE
# ==================================================

# Ciudades simuladas
cities = {
    "BOG": "Bogotá",
    "MED": "Medellín",
    "CAL": "Cali",
    "BAR": "Barranquilla",
    "CAR": "Cartagena",
    "BUC": "Bucaramanga",
    "PER": "Pereira",
    "MAN": "Manizales",
    "CUC": "Cúcuta",
    "IBA": "Ibagué"
}

# Zonas simuladas
zones = [
    "Centro", "Norte", "Sur", "Occidente", "Oriente",
    "Poblado", "Usaquen", "Chapinero", "Calima",
    "Mirandela", "LaFlora", "SanFernando",
    "Castilla", "Kennedy", "Suba", "Envigado", "Belen"
]

# Lista para almacenar estaciones base
bs_data = []

# Identificador inicial
bs_id = 1

# Generación de estaciones base
for code, city in cities.items():

    for _ in range(10):

        bs_data.append({
            "bs_id": bs_id,
            "bs_name": f"{code}.{random.choice(zones)}",
            "city": city
        })

        bs_id += 1

# Crear DataFrame
bs_df = pd.DataFrame(bs_data)

# Mostrar primeras filas
bs_df.head()
````

## ✅ Análisis

Se generó exitosamente la infraestructura básica de estaciones base para la red móvil simulada.

Las estaciones fueron distribuidas entre diferentes ciudades principales de Colombia, permitiendo representar un entorno geográfico más realista para el análisis de tráfico y desempeño de red.

Cada ciudad cuenta con múltiples estaciones base ubicadas en diferentes zonas, lo que facilitará posteriormente:

- Simulación de tráfico regional

# 📞 Generación de Registros de Llamadas (CDRs)

En esta sección se generarán los registros de llamadas o **CDRs (Call Detail Records)**, que representan el núcleo principal del proyecto de análisis de telecomunicaciones.

Cada registro de llamada almacenará información relevante sobre el comportamiento de la red y los usuarios móviles, incluyendo:

- Línea móvil asociada
- Número de destino
- Tipo de llamada
- Fecha y hora de inicio
- Duración
- Costo
- Estación base origen y destino
- Estado de la llamada
- Tecnología utilizada (3G, 4G o 5G)

Adicionalmente, se implementarán reglas probabilísticas para simular condiciones reales de red, donde la calidad del servicio dependerá parcialmente de la tecnología utilizada.

Esto permitirá generar escenarios más cercanos a una operación real de telecomunicaciones y facilitará posteriormente el análisis de KPIs de calidad y desempeño.

---

````python
# ==================================================
# CONFIGURACIÓN DE REGISTROS DE LLAMADAS
# ==================================================

# Lista para almacenar registros
call_data = []

# Estados posibles de llamada
call_statuses = [
    'completed',
    'congestion',
    'dropped',
    'no_signal',
    'no_answer'
]

# Tipos de llamada
call_types = [
    'outgoing',
    'incoming',
    'missed'
]

# Tecnologías móviles
technologies = ['3G', '4G', '5G']

# Distribución de tecnologías
tech_weights = [0.15, 0.60, 0.25]

# Identificador inicial
call_id = 1

print("✅ Configuración inicial de CDRs completada")
````

## ✅ Análisis

Se definieron las variables principales necesarias para la generación de los registros de llamadas.

El modelo contempla diferentes estados de llamada y tecnologías móviles, permitiendo simular múltiples escenarios operativos de una red de telecomunicaciones.

La distribución de tecnologías fue configurada para representar un entorno moderno de red móvil:

- 4G como tecnología predominante
- 5G en crecimiento
- 3G con menor participación

Estas configuraciones impactarán directamente los indicadores de calidad de servicio y desempeño de red generados posteriormente.

---

````python
# ==================================================
# GENERACIÓN DE REGISTROS DE LLAMADAS (CDRs)
# ==================================================

for line_id, profile in line_profiles.items():

    # Generación de volumen de llamadas según perfil
    if profile == 'high':
        num_calls = random.randint(500, 1200)

    elif profile == 'medium':
        num_calls = random.randint(100, 400)

    else:
        num_calls = random.randint(0, 50)

    # Generación de llamadas
    for _ in range(num_calls):

        # Duración inicial
        duration = random.randint(0, 600)

        # Fecha aleatoria dentro de los últimos 30 días
        start = datetime.now() - timedelta(days=random.randint(0, 30))

        # Hora final
        end = start + timedelta(seconds=duration)

        # Costo estimado
        cost = round(duration * 0.02, 2)

        # ==========================================
        # TECNOLOGÍA
        # ==========================================

        technology = random.choices(
            technologies,
            weights=tech_weights
        )[0]

        # ==========================================
        # ESTADO DE LLAMADA SEGÚN TECNOLOGÍA
        # ==========================================

        if technology == '3G':

            status_weights = [0.60, 0.15, 0.10, 0.10, 0.05]

        elif technology == '4G':

            status_weights = [0.75, 0.10, 0.07, 0.04, 0.04]

        else:  # 5G

            status_weights = [0.85, 0.07, 0.04, 0.02, 0.02]

        # Estado final de llamada
        status = random.choices(
            call_statuses,
            weights=status_weights
        )[0]

        # Ajuste de duración para llamadas fallidas
        if status != 'completed':

            duration = random.randint(0, 10)

            cost = round(duration * 0.02, 2)

        # Registro final
        call_data.append({

            "call_id": call_id,
            "line_id": line_id,

            "destination_number":
                f'3{random.randint(100000000,999999999)}',

            "call_type": random.choice(call_types),

            "start_time": start,
            "end_time": end,

            "duration_seconds": duration,
            "cost": cost,

            "origin_bs_id": random.randint(1, 100),
            "destination_bs_id": random.randint(1, 100),

            "call_status": status,
            "technology": technology
        })

        call_id += 1

# Crear DataFrame
calls_df = pd.DataFrame(call_data)

print("✅ Registros de llamadas generados correctamente")
print(f"📞 Total de llamadas generadas: {len(calls_df):,}")
````

## 📊 Análisis de Generación de CDRs

Se generaron exitosamente los registros de llamadas (CDRs) que representan el tráfico principal de la red móvil simulada.

El modelo implementado permite representar múltiples características presentes en entornos reales de telecomunicaciones:

- Diferencias de consumo entre usuarios
- Distribución desigual de tráfico
- Variaciones de calidad según tecnología
- Eventos de congestión y fallas
- Comportamiento probabilístico de la red

Adicionalmente, las tecnologías más modernas presentan mayores probabilidades de llamadas exitosas, simulando mejoras reales en cobertura, estabilidad y desempeño de red.

# 📈 KPIs de Telecomunicaciones

En esta sección se calcularán algunos de los principales KPIs (Key Performance Indicators) utilizados en la industria de telecomunicaciones para evaluar el desempeño y la calidad de servicio de una red móvil.

Los indicadores permitirán medir aspectos como:

- Éxito de llamadas
- Congestión de red
- Calidad del servicio
- Estabilidad de comunicación
- Desempeño tecnológico

El análisis de KPIs es fundamental en operadores móviles para monitorear la experiencia de usuario y detectar posibles problemas operativos dentro de la infraestructura de red.

---

# 📞 KPI: CSSR (Call Setup Success Rate)

El CSSR representa el porcentaje de llamadas que fueron establecidas exitosamente dentro de la red móvil.

Este indicador es uno de los KPIs más importantes en telecomunicaciones, ya que permite evaluar la capacidad de la red para completar correctamente las solicitudes de llamada realizadas por los usuarios.

La fórmula general del CSSR es:

CSSR=( Llamadas Completadas / Total de Llamadas) ​×100


````python
# ==================================================
# KPI - CSSR
# ==================================================

# Total de llamadas
total_calls = len(calls_df)

# Llamadas completadas
completed_calls = len(
    calls_df[calls_df['call_status'] == 'completed']
)

# Cálculo CSSR
cssr = round(
    (completed_calls / total_calls) * 100,
    2
)

print(f"CSSR: {cssr}%")
````

## 📊 Análisis del CSSR

El indicador CSSR obtenido fue de **75.21%**, lo que significa que aproximadamente tres de cada cuatro llamadas fueron completadas exitosamente dentro de la red móvil simulada.

Este resultado refleja un entorno de red con un nivel aceptable de desempeño, aunque aún se presentan eventos de falla asociados a congestión, pérdida de señal y degradación tecnológica.

La configuración probabilística implementada durante la generación de datos permite representar escenarios más cercanos a una operación real, donde no todas las llamadas logran establecerse correctamente.

Adicionalmente, las tecnologías 4G y 5G contribuyen positivamente al indicador debido a sus mayores probabilidades de éxito, mientras que 3G presenta un comportamiento más limitado en términos de estabilidad y calidad de servicio.

En entornos reales de telecomunicaciones, el CSSR es uno de los KPIs más importantes para evaluar la capacidad operativa de la red y la experiencia general del usuario.

---

# 📉 KPI: DCR (Dropped Call Rate)

El DCR representa el porcentaje de llamadas que fueron interrumpidas o finalizadas inesperadamente durante la comunicación.

Este KPI es utilizado para medir la estabilidad de la red y detectar problemas relacionados con:

- Cobertura
- Calidad de señal
- Interferencias
- Movilidad
- Congestión

DCR = (Llamadas Caídas / Total de Llamadas) × 100


````python
# ==================================================
# KPI - DCR
# ==================================================

# Llamadas caídas
dropped_calls = len(
    calls_df[calls_df['call_status'] == 'dropped']
)

# Cálculo DCR
dcr = round(
    (dropped_calls / total_calls) * 100,
    2
)

print(f"DCR: {dcr}%")
````

## 📊 Análisis del DCR

El indicador DCR obtenido fue de **6.71%**, lo que representa el porcentaje de llamadas que fueron interrumpidas durante la comunicación dentro de la red móvil simulada.

Este resultado evidencia la presencia de eventos de degradación de servicio asociados a factores como congestión, pérdida de señal y limitaciones tecnológicas.

Las tecnologías más modernas, como 4G y 5G, contribuyen a mantener niveles más bajos de llamadas caídas, mientras que 3G presenta una mayor probabilidad relativa de interrupciones debido a sus limitaciones de capacidad y estabilidad.

En entornos reales de telecomunicaciones, mantener un DCR bajo es fundamental para garantizar una experiencia de usuario estable y minimizar afectaciones en la continuidad del servicio.

---

# 🚦 KPI: Congestion Rate

El Congestion Rate representa el porcentaje de llamadas afectadas por congestión dentro de la red móvil.

La congestión ocurre cuando la capacidad disponible de la infraestructura no es suficiente para atender la demanda de tráfico generada por los usuarios.

### Fórmula del Congestion Rate

```text
Congestion Rate = (Llamadas con Congestión / Total de Llamadas) × 100
```

---

````python
# ==================================================
# KPI - CONGESTION RATE
# ==================================================

# Llamadas con congestión
congestion_calls = len(
    calls_df[calls_df['call_status'] == 'congestion']
)

# Cálculo de congestión
congestion_rate = round(
    (congestion_calls / total_calls) * 100,
    2
)

print(f"Congestion Rate: {congestion_rate}%")
````

## 📊 Análisis de Congestión

El indicador de congestión obtenido fue de **10.11%**, lo que indica que aproximadamente una de cada diez llamadas presentó afectaciones relacionadas con saturación de red.

Este comportamiento simula escenarios reales donde la infraestructura móvil puede experimentar limitaciones de capacidad debido a incrementos de tráfico, alta demanda de usuarios o restricciones tecnológicas.

Las tecnologías más modernas presentan menores niveles relativos de congestión gracias a una mayor capacidad y eficiencia operativa, mientras que tecnologías heredadas como 3G tienden a verse más afectadas en escenarios de alta utilización.

En entornos reales de telecomunicaciones, este KPI es monitoreado constantemente para identificar necesidades de expansión, optimización de red y mejoras de cobertura.

````python
# ==================================================
# CREACIÓN DE DÍA DE LA SEMANA
# ==================================================

# Extraer día de la semana
calls_df['day_name'] = calls_df['start_time'].dt.day_name()

# Vista rápida
calls_df[['start_time', 'day_name']].head()
````

# 🔗 Consolidación de Datos

En esta sección se consolidarán los diferentes datasets generados durante el proyecto en un único archivo enriquecido.

El objetivo es integrar:

- Información de llamadas
- Información de usuarios
- Información de estaciones base

Esto permitirá construir un dataset más completo y optimizado para análisis y visualización en Power BI.

El archivo final generado será:

```text
calls_enriched.csv
```

---

````python
# ==================================================
# UNIÓN DE LÍNEAS MÓVILES
# ==================================================

# Unir información de usuarios
calls_enriched = calls_df.merge(
    lines_df[
        [
            'line_id',
            'phone_number',
            'user_name',
            'department'
        ]
    ],
    on='line_id',
    how='left'
)

# Validación
calls_enriched.head()
````

````python
# ==================================================
# UNIÓN DE ESTACIONES BASE ORIGEN
# ==================================================

# Información de estación origen
origin_bs = bs_df.rename(columns={
    'bs_id': 'origin_bs_id',
    'bs_name': 'origin_bs_name',
    'city': 'origin_city'
})

# Merge origen
calls_enriched = calls_enriched.merge(
    origin_bs[
        [
            'origin_bs_id',
            'origin_bs_name',
            'origin_city'
        ]
    ],
    on='origin_bs_id',
    how='left'
)

# Validación
calls_enriched.head()
````

````python
# ==================================================
# UNIÓN DE ESTACIONES BASE DESTINO
# ==================================================

# Información de estación destino
destination_bs = bs_df.rename(columns={
    'bs_id': 'destination_bs_id',
    'bs_name': 'destination_bs_name',
    'city': 'destination_city'
})

# Merge destino
calls_enriched = calls_enriched.merge(
    destination_bs[
        [
            'destination_bs_id',
            'destination_bs_name',
            'destination_city'
        ]
    ],
    on='destination_bs_id',
    how='left'
)

# Validación
calls_enriched.head()
````

````python
# ==================================================
# SELECCIÓN FINAL DE COLUMNAS
# ==================================================

calls_enriched = calls_enriched[
    [
        'call_id',
        'line_id',
        'destination_number',
        'call_type',
        'start_time',
        'end_time',
        'duration_seconds',
        'cost',
        'origin_bs_id',
        'destination_bs_id',
        'call_status',
        'technology',
        'phone_number',
        'user_name',
        'department',
        'origin_bs_name',
        'origin_city',
        'destination_bs_name',
        'destination_city'
    ]
]

# Validación final
calls_enriched.head()
````

````python
# ==================================================
# EXPORTACIÓN FINAL
# ==================================================

calls_enriched.to_csv(
    "calls_enriched.csv",
    index=False
)

print("✅ Archivo calls_enriched_v2.csv generado correctamente")
````
